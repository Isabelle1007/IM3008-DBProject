```python
import urllib.request as req
import json
import pandas as pd
```

## courselist

```python
courses = pd.DataFrame(columns = ['c_id', 'c_name', 'w_id', 'price', 'total_time', 't_id', 'start_time', 'type', 'buyers', 'raisefunding', 'cg_id'])
teachers = pd.DataFrame(columns = ['t_id', 't_name'])
website = pd.DataFrame(columns = ['w_id', 'website_name'])
categories = pd.DataFrame(columns = ['cg_id', 'cg_name'])

for i in range(29):
    url = "https://api.hahow.in/api/courses?limit=24&page=" + str(i)
    request = req.Request(url, headers = {
        "Content-Type":"application/json; charset=utf-8",
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
    })
    with req.urlopen(request) as response:
        result = response.read().decode("utf-8")
    result = json.loads(result)
    for j in range(len(result['data'])):
        #courses
        data_courses = {'c_id':result['data'][j]['_id'], 'c_name':result['data'][j]['title'], 'w_id':result['data'][j]['_id']
                , 'price':result['data'][j]['price']
                , 't_id':result['data'][j]['owner']['_id'], 'buyers':result['data'][j]['numSoldTickets'], 'raisefunding':result['data'][j]['status']}
        courses = courses.append(data_courses, ignore_index = True)
        #teahcers
        if result['data'][j]['owner']['_id'] not in teachers['t_id']:
            data_teachers = {'t_id':result['data'][j]['owner']['_id'], 't_name':result['data'][j]['owner']['name']}
            teachers = teachers.append(data_teachers, ignore_index = True)
        #website
        data_website = {'w_id':result['data'][j]['_id'], 'website_name':"https://hahow.in/courses/" + result['data'][j]['_id']}
        website = website.append(data_website, ignore_index = True)
courses
```

```python
url = "https://api.hahow.in/api/groups"
request = req.Request(url, headers = {
        "Content-Type":"application/json; charset=utf-8",
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
    })
with req.urlopen(request) as response:
    result = response.read().decode("utf-8")
result = json.loads(result)

for i in range(len(result)):
    data_categories = {'cg_id':result[i]['_id'], 'cg_name':result[i]['title']}
    categories = categories.append(data_categories, ignore_index = True)
categories
```
