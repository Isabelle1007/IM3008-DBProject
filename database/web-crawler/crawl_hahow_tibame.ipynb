## courselist

```python
import urllib.request as req
import json
import pandas as pd

courses = pd.DataFrame(columns = ['c_id', 'c_name', 'w_id', 'price', 'total_time', 't_id', 'start_time', 'type', 'buyers', 'raisefunding', 'cg_id'])

for i in range(29):
    url = "https://api.hahow.in/api/courses?limit=24&page=" + str(i)
    request = req.Request(url, headers = {
        "Content-Type":"application/json; charset=utf-8",
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
    })
    with req.urlopen(request) as response:
        result = response.read().decode("utf-8")
    result = json.loads(result)
    for j in range(len(result['data'])):
        data = {'c_id':result['data'][j]['_id'], 'c_name':result['data'][j]['title'], 'w_id':1, 'price':result['data'][j]['price']
                , 't_id':result['data'][j]['owner']['_id'], 'buyers':result['data'][j]['numSoldTickets'], 'raisefunding':result['data'][j]['status']}
        courses = courses.append(data, ignore_index = True)
courses
```

# Tibame

```python
import urllib.request as req
import json
import pandas as pd
import requests

url = "https://api.tibame.com/courselib/user/courseList"
requestData = {"orderBy":"popular","sortType":"DESC","isPublic":1,"numPerPage":18,"isCategory":False,"isSubCategory":False}
request = req.Request(url, headers = {
    "Content-Type":"application/json; charset=utf-8",
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
}, data = json.dumps(requestData).encode("utf-8"))

with req.urlopen(request) as response:
    result = response.read().decode("utf-8")
print(result)

result = json.loads(result)
print(result['data']['results'][0]['courseName']) 
```
